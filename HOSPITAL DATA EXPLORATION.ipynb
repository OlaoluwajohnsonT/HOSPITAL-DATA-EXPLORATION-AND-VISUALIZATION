All the columns are integer and I need to covert them to float and to do that, 
I will need to remove the dollar sign fromeach columns using a function

```python
def remove_dols(d):
    d = d.replace('$','')
    d = d.replace(',','')
    d = d.replace(' ','')
    d = d.replace('-','')
    d = pd.to_numeric(d)
    return d

df['Medication Revenue'] = df['Medication Revenue'].apply(remove_dols)
df['Lab Cost'] = df['Lab Cost'].apply(remove_dols)
df['Consultation Revenue'] = df['Consultation Revenue'].apply(remove_dols)
```

# Now let see the time that has the most visitor

Using groupby, we can find the time has the most visitor

```python
df_visitor= df.groupby(df['Entry Hour'])['Patient ID'].count()
df_visitor
#We can see that 9 am has the most visitor hour
```

```python
plt.style.use('ggplot')
df_visitor= df.groupby(df['Entry Hour'])['Patient ID'].count().plot()

plt.title('PATIENT BY ENTRY HOUR')
```

# So let check if the total revenue correlate with the total visitor

```python
df_revenue= df.groupby(df['Entry Hour'])['Total Revenue'].sum().plot()
plt.title('TOTAL REVENUE')
```

# LET PLOT BOTH GRAPH ON THE SAME AXIS

```python
df_hour= df.groupby(df['Entry Hour']).count()
df_sum= df.groupby(df['Entry Hour']).sum()
```

```python
merge=df_sum.merge(df_hour, on='Entry Hour')
merge.head()
```

```python
fig, ax1 = plt.subplots()
ax2= ax1.twinx()
ax1.plot(merge.index, merge['Total Revenue_x'], color ='b')
ax2.plot(merge.index, merge['Patient ID'], color='r')
plt.title('PATIENT ID AND TOTAL REVENUE CORRELATION')
```

```python
fig, ax1 = plt.subplots()
ax2= ax1.twinx()
ax1.bar(merge.index, merge['Total Revenue_x'], color ='b')
ax2.plot(merge.index, merge['Patient ID'], color='r')
plt.title('PATIENT ID AND TOTAL REVENUE CORRELATION')
```
